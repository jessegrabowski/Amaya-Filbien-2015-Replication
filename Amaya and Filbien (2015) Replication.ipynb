```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re

from typing import Optional, Callable
from collections import defaultdict 

from pandas_datareader import fred

from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.lm.preprocessing import everygrams
from string import punctuation
from unidecode import unidecode_expect_ascii

from scipy import signal
import statsmodels.api as sm

plot_defaults = {
    'figure.figsize':(14,4),
    'figure.facecolor':'white',
    'axes.grid':True,
    'grid.linewidth':0.5,
    'grid.linestyle':'--',
    'axes.spines.top':False,
    'axes.spines.bottom':False,
    'axes.spines.left':False,
    'axes.spines.right':False
}

plt.rcParams.update(plot_defaults)
```

```python
def make_sentiment_dict(pos_words:list[str], neg_words:list[str]) -> dict[str, float]:
    sent_dict = defaultdict(lambda: 0)
    sent_dict.update({w.lower(): 1 for w in POS_WORDS})
    sent_dict.update({w.lower():-1 for w in NEG_WORDS})
    
    return sent_dict

def preprocess_text(s: str, stemmer: Optional[Callable]=None) -> list[str]:
    '''
    Prepare a text string for sentiment analysis
    
    Parameters
    ----------
    s: str,
        The string to be preprocessed
        
    stemmer: callable, Optional
        A stemming function to be applied to each token in the string. Should
        take a string and return a string.
        
    Returns
    ----------
    tokens: list of str,
        A list of tokens, lowercased and stemmed.
    '''
    
    if not stemmer:
        stemmer = lambda x: x
    
    s = s.lower()
    tokens = word_tokenize(s)
    stemmed_tokens = [stemmer(w) for w in tokens]
    
    return stemmed_tokens

def classify_sentences(corpus: list[str], 
                       pos_words: list[str],
                       neg_words: list[str],
                       stemmer: Optional[Callable]=None,
                       average_method:str='all') -> list[float]:
    """
    Classify the sentiment of each text in a corpus based on a fixed dictionary.
    
    Parameters
    -----------
    corpus: list of strings
        A "corpus" of texts. Each text will be converted to a single number, 
        representing the sentiment of the text.
    
    pos_words: list of str
        A list of words to be considered as positive in the sentiment
        analysis
        
    neg_words: list of str
        A list of words to be considered as negative in the sentiment 
        analysis.
    
    stemmer: callable, optional
        A stemming function to be applied to each token in the string. Should
        take a string and return a string.
        
    average_method: str, default = 'all'
        The method used to normalize the sentiment scores between 0 and 1. 
            "all" will divide the sentiment by the total number of tokens in the text
            "has_sentiment" will divide by the number of non-zero sentiment tokens
    
    Returns
    ----------
    scores: list of floats
        A list of sentiment scores
    
    """
    if average_method not in ['all', 'has_sentiment']:
        raise NotImplementedError()
    
    sent_dict = make_sentiment_dict(pos_words, neg_words)
        
    scores = []
    for text in corpus:
        tokens = preprocess_text(text, stemmer)
        word_scores = list([sent_dict[token] for token in tokens])
        n = len(tokens) if average_method == 'all' else sum([abs(x) for x in word_scores])
        score = sum(word_scores) / n
        scores.append(score)
    return scores
```

# Clean Text

```python
junk_strings = ['Click here for the transcript of questions and answers',
               'Jump to the transcript of the questions and answers',
                'With the transcript of the questions and answers',
                'With a transcript of the questions and answers']
junk_strings = [x.strip().lower() for x in junk_strings]
```

```python
def clean_text(row):
    text = row.text
    date = row.date
    location = '' if pd.isna(row.location) else row.location 
    lines = text.split('\n')
    
    result = []
    
    for line in lines:
        if date in line.strip() and location in line.strip():
            continue
        elif re.sub('[^\s\d\w]', '', line.strip().lower()) in junk_strings:
            continue
        result.append(line)
    return '\n'.join(result)
```

```python
stemmer = PorterStemmer().stem
PUNCT_PATTERN = re.compile('[' + re.escape(punctuation.replace("'", '')  + '\n') + '\d' + ']')
stop_words = [stemmer(word_tokenize(x)[0]) for x in stopwords.words('english')]

def process_text(s):
    s = s.lower()
    s = re.sub(PUNCT_PATTERN, ' ', s)
    s = re.sub(' +', ' ', s)
    s = unidecode_expect_ascii(s)
    tokens = word_tokenize(s)
    stemmed_tokens = [stemmer(x) for x in tokens if x not in stop_words]
    bigrams = everygrams(stemmed_tokens, min_len=2, max_len=2)
    
    return set(bigrams)
```

```python
df_bigrams = df.intro_text.apply(process_text)
```

```python
def jaccard_similarity(a, b):
    union = a.union(b)
    intersection = a.intersection(b)
    
    return len(intersection) / len(union)

N = df.shape[0]
similarity_matrix = np.eye(N)
for i in range(N):
    for j in range(i):
        similarity_matrix[i, j] = jaccard_similarity(df_bigrams.iloc[i], df_bigrams.iloc[j])
```

```python
L1_similarity = pd.Series(np.diag(similarity_matrix, k=-1), index=df.index[1:], name='Similarity')
L1_similarity.plot()
```

## Equation 4

$$ \log{\text {Similarity}_i} = \alpha_0 + \alpha_1 \log{\text{Time}} + X\beta + \epsilon_i$$

Log Time is a new one for me.

```python
def safe_log(x):
    out = np.log(x + 1e-8)
    return np.maximum(0, out)

eq4_df = data_df.assign(log_Similarity = lambda x: np.log(x.Similarity),
                        Time = lambda x: safe_log((x.index - x.index[0]).days),
                        log_Time = lambda x: safe_log(x.Time))
                        
eq1 = 'log_Similarity ~ 1 + output_gap + inflation + delta_MRO'
eq2 = 'log_Similarity ~ 1 + log_Time'
eq3 = 'log_Similarity ~ 1 + log_Time + output_gap + inflation + delta_MRO'
eq4 = 'log_Similarity ~ 1 + Time + output_gap + inflation + delta_MRO'

regressions_1 = []
for formula in [eq1, eq2, eq3, eq4]:
    mod = sm.OLS.from_formula(formula, eq4_df)
    res = mod.fit(cov_type='HC0')
    regressions_1.append(res)
```

### Table 3

Results are qualitatively similar to those in the paper: time matters, the output gap matters, while delta MRO doesn't. In my results, inflation is not robust to the inclusion of time, which is actually the opposite of what the paper finds. These results are intersting in light of discussion about the charter of the ECB, which is offically has only a single objective -- price stability. Nevertheless, if we believe this similarity measure, bank communication is more influenced by the output gap, which is not an official target, than the official target of inflation.

Looking at the graphs I made above, there are likely to be stationarity issues in the independent variables, so there will be confounding between them and time. There is also confounding between inflation and the output gap, since inflation is linked to interest rates, and interest rates are linked to the output gap viavia the central bank's Taylor rule (I don't believe $\beta = 0$ for the ECB). So I would say this regression is not identified.

```python
from stargazer.stargazer import Stargazer
from IPython.display import HTML
sg_output = Stargazer(regressions_1)
HTML(sg_output.render_html())
```

## Equation 9

$$ |CAR_i | = \gamma_0 + \gamma_1 \log{\text{Similarity}_i} \cdot \text{Pessimism}_i + X\beta + \epsilon_i$$

```python
eq1 = 'abs_CAR ~ 1 + Pessimism'
eq2 = 'abs_CAR ~ 1 + output_gap + delta_MRO'
eq3 = 'abs_CAR ~ 1 + Pessimism * log_Similarity - Pessimism - log_Similarity'
eq4 = 'abs_CAR ~ 1 + Pessimism * log_Similarity - Pessimism - log_Similarity + output_gap + inflation + delta_MRO'

regressions = []
for formula in [eq1, eq2, eq3, eq4]:
    mod = sm.OLS.from_formula(formula, data_df.rename(columns={'intro_sentiment':'Pessimism'}).assign(log_Similarity = lambda x: np.log(x.Similarity)))
    res = mod.fit(cov_type='HC0')
    regressions.append(res)
```

### Table 4

I've omitted the MRO surprise variable because I don't have access to historial analyist consensus data, and it's not reported as significant anyway.

Nevertheless, results aren't so far off from the paper. Pessimism is associated with CAR iwtha  coefficient of about 0.5 in regression (1), but the relation becomes significant and negative when it's interacted with log similarity. Dropping the log_similarity term itself and only leaving the interaction term is a bit suss, as the kids say.

As in the paper, none of the macro controls are statistically significant.

```python
sg_output = Stargazer(regressions)
HTML(sg_output.render_html())
```

## Bayesian Model Comparison

Setting aside the president effect, we can re-run all the models presented in the paper and use Bayesian model averaging to see which one best explains the data.

```python
def make_model(y_df, X_df=None, time=None):
    include_time = time is not None
    include_controls = X_df is not None
    coords = {'variables':X_df.columns} if include_controls else None

    with pm.Model(coords=coords) as model:
        y = pm.Data('y', y_df, mutable=False)

        intercept = pm.Normal('intercept', mu=-1.5, sigma=1)
        mu = intercept
        
        if include_controls:
            X = pm.Data('X', X_df, mutable=False)
            betas = pm.Normal('betas', mu=0, sigma=1, dims=['variables'])
            mu += X @ betas
            
        if include_time:
            t = pm.Data('t', time, mutable=False)
            trend = pm.Normal('trend', mu=0, sigma=1)
            mu += t * trend
        
        sigma = pm.HalfNormal('sigma', sigma=1)
        y_hat = pm.Normal('y_hat', mu=mu, sigma=sigma, observed=y)

    return model
```

# Extension 2: Perplexity as an alternative measure of similarity

People don't actually read documents as a collection of bigrams. People read documents like a time series, one word at a time, and integrate information as they go. As we read, we are trying to decode the meaning of the document. A well written document quickly allows the reader to decode this latent "thought vector", which then helps you to predict the next words, and more easily absorb the information.

A measure that comes closer to capturing this natural method of reading and information integration is preplexity. If we have a generative, probabilistic language model of the form:

$$P \left ( w_{t+1} | \{w_\tau\}_{\tau=0}^t \right )$$

Given this model, we can evaluate how "surprised" we are to see a given sequence of words, $\{w_t\}$, by computing the entropy of the sequence, then exponentiating:

$$ \begin{align} PP(\{w_t\}) &= 2^{H(\{w_t\})}, \quad \text{with} \\
                          H(\{w_t\}) &= -\frac{1}{N}\log_2(p(\{w_t\}) \end{align}$$

Taking the average perplexity per word of a document has a nice natural interpertation. If a sentence has a perplexity of 6, we are as confused by the next word as if we had to choose uniformly between 6 words. Thus, the perplexity score is bounded between 1 and the number of words in the vocabulary.

## Simple Bigram Model

To implement this method, we need a generative probabilistic model. The simplest is a bigram model, which `nltk` can do for us. We'll impose a Markov structure on the probability, so that only the previous $s$ words determine the probability of seeing the next word:

$$P \left ( w_{t+1} | \{w_{t-\tau} \}_{\tau=1}^{s-1} \right )$$

Common choices of $s$ are $s=2$, producing the bigram model:

$$ P \left ( w_{t+1} | w_t \right ) $$

And $s=3$, producing the trigram model:

$$ P \left ( w_{t+1} | w_t, w_{t-1} \right ) $$

Using this formulation the probability of a sentence or text can be evaluated as the product of the whole sequence. Taking $s=1$ (the unigram model), the probablity of a sentence will be the joint probability of observing all the words in that sentence, or:

$$ P (\{w_t\}) = p(w_0)^{c(w_0)} \cdot p(w_1)^{c(w_1)} \cdot \dots \cdot p(w_t)^{c(w_t)}$$

Where $c(w_t)$ is a counting function that returns the number of times a word appears in the sentence. This can be solved as a constrained optimization problem without imposing distributional assumptions on $p(w_t)$, because we have the constraint $\sum_{\tau=0}^t p(w_\tau) = 1$. Noting that the log function is monotonic, we maximize log probability instead of probability, and form the following Lagrangian can be formed:

$$\mathcal{L} = \sum_{\tau=0}^t c(w_\tau) \log{p(w_\tau)} - \lambda \left ( \sum_{\tau=0}^t p(w_\tau) - 1 \right )$$

Maximizing this with respect to $p(w_\tau)$ and $\lambda$ gives:

$$\begin{align} p(w_{\tau}) &= -\frac{c(w_\tau)}{\lambda} \\
                  \lambda &= \sum_{\tau=0}^t c(w_\tau) \end{align}$$
                  
So that:

$$p(w_{\tau}) = \frac{c(w_\tau)}{\sum_{\tau=0}^t c(w_\tau)}$$

The probability MLE estimate of a word in the unigram model is simply the number of times it appears in a corpus, divided by the size of the corpus. This general procedure remains true for the bigram and trigram models, and can be generalized by re-writing the corpus as "words" of bigrams, trigrams, or ngrams. This is possible because of the exchangability of words within a sentence induced by the markov assumption. In the unigram model, the probability of "the cow jumped over the moon" is equivalent to "cow the moon over jumped", since all sequence information is lost. In a bigram model, this sentence can be considered as a sequence of single bigram "words": "(the cow) (cow jumped) (jumped over) (over the) (the moon)", and these new "words" are once again exchangable, and the MLE estimator derived above will apply.

```python
import spacy
from nltk.tokenize import wordpunct_tokenize
from collections import deque
from unidecode import unidecode
nlp = spacy.load("en_core_web_sm")
KEEP_PUNCT = '''.?!'"$%'''
PUNCT = ''.join([x for x in punctuation if x not in KEEP_PUNCT])

def clean_text(s):
    lower = s.lower()
    decoded  = unidecode_expect_ascii(lower)
    no_punct = re.sub('[' + re.escape(PUNCT) + '\n\t' + ']', ' ', decoded)
    pad_numbers = re.sub('\d', ' \g<0> ', no_punct)
    reduce_spaces = re.sub(' +', ' ', pad_numbers)
    return reduce_spaces

def generate_sentence(model, n):
    sentence = []
    last_words = deque(['<s>'] * n, n)
    next_word = None
    while '</s>' not in last_words:
        next_word = model.generate(1, text_seed=last_words)
        sentence.append(next_word)
        last_words.append(next_word)
    
    sentence = ' '.join(s for s in sentence[:-1])
    sentence = re.sub(f'( ({KEEP_PUNCT}) ?)', '\g<2> ', sentence)
    sentence = re.sub('( ?(\d) ?)', '\g<2>', sentence)
    sentence = re.sub('\$?\d+\.?,?%?\d*', ' \g<0> ', sentence)
    sentence = re.sub(' +', ' ', sentence)
    return sentence.strip()


texts = df.intro_text.copy()
texts = texts.apply(clean_text)
```

## Perplexity Patch 1: Bayesian Learning

So there are two problems with the MLE model approach to perplexity. The first is that ngrams are exchangable, so it doesn't actually accomplish what I hoped it would at the outset: model how people read documents as a time series. In the second, there is severe information leakage across time, which ruins the approach as a metric for later econometric estimation.

These can both be fixed. The problem of information leakage can be resolved by re-fitting the language model every time a new documnent is issued by the bank. The actual learning procedure people go through is as follows:

1. Live for 20+ years, accumulating a knowledge base about the world, potentially including specialized information about finance and central banking.
2. Read an ECB document for the first time. Your expectations while reading are informed by (1).
3. Integrate the information from the document into your pool of knowledge.
4. Read another ECB document. Expectations while reading are not informed by the first document, plus life experience.

And so on.

This learning process is captured by Bayesian updating. Rather than eschewing distributional assumptions about word probabilities $p(w_t)$ (where $w$ is some ngram, not necessarily a single word), we assume that words are Categorically distributed, so that each of $K$ words in our vocabulary has some probabiltiy $p$ of being randomly sampled. We further assume that the reader's prior distribution over the same $K$ words is Dirichlet distributed, with a weighting parameter $\alpha$ that summarizes relative frequencies between words.

This setup allows for conjugate updating of the categorical word distribution one document at a time. The posterior value of each $\alpha_i$ in the vector $\alpha$ will simply be $\alpha_i + c(w_i)$, while the posterior distribution itself will be categorically distributed with $p = \frac{\alpha_i + c(w_i)}{\sum_i (\alpha_i + c(w_i))}$

This also gives intuition to $\alpha$: it is a running tally of how many times a word has been seen in the reader's life. 

Note that this setup is precisely the MLE setup, but regularized by the prior parameter $\alpha$.

The procedure will now go as follows:

1. Obtain a prior estimate for $\alpha$.
2. Set p = $\frac{\alpha_i}{\sum_i \alpha_i}$, compute the perplexity of the first document.
3. Update $\alpha_i = \alpha_i + c(w_i)$ and $p = \frac{\alpha_i + c(w_i)}{\sum_i (\alpha_i + c(w_i))}$ using the words in the first document.
4. Repeat.

### Choice of prior $\alpha$

The lazy choice would be to set the probability of all words as being equal, but we can do better with just a bit of legwork. I found a book from 1992, "Fiscal Policy, Taxation and the Financial System in an Increasingly Integrated Europe", and extracted unigram and bigram counts from it. This should represent a reasonable corpus about central bank financial issues in the European Union circa 1990. It will be a bit out of date, as the ECB had not yet been established, but it should serve as a reasonable prior over English words generally, with a tilt towards finance and economics. The model will have to "learn" the correct probabilities for ECB specific words itself.

```python
prior_unigrams = pd.read_csv('data/unigram_counts.csv', index_col=0)
prior_bigrams = pd.read_csv('data/bigram_counts.csv', index_col=[0, 1])
```

```python
prior_unigrams.sort_values(by='Unigrams', ascending=False, inplace=True)
prior_unigrams['rank'] = prior_unigrams.Unigrams.rank(ascending=False)
```

```python
prior_bigrams.sort_values(by='Bigrams', ascending=False, inplace=True)
prior_bigrams['rank'] = prior_bigrams.Bigrams.rank(ascending=False)
```

The last detail to handle is unknown words. We can use a `defaultdict` to assign a value of $\alpha = 1$ to all possible bigrams by default.

```python
from collections import defaultdict
from collections import Counter
alpha = defaultdict(lambda: 1)
alpha.update(prior_bigrams.Bigrams.to_dict())
```

```python
def process_document(doc):
    clean_doc = clean_text(doc)
    sentences = [word_tokenize(x.text) for x in nlp(clean_doc).sents]
    bigrams_iter_obj , _ = padded_everygram_pipeline(2, sentences)
    bigrams = [s for x in list(bigrams_iter_obj) for s in list(x) if len(s) == 2]

    return bigrams

def compute_perplexity(bigrams, doc_counts, alpha):
    probs = []
    
    denom = sum([doc_counts[w] + alpha[w] for w in list(set(bigrams))])
    for bigram in bigrams:
        p = (doc_counts[bigram] + alpha[bigram]) / denom
        probs.append(p)
    probs = np.array(probs)
    H = -(np.log2(probs)).mean()
    PP = 2 ** H
    
    return PP

def update_alpha(bigrams, alpha):
    for bigram in bigrams:
        alpha[bigram] += 1
    
    return alpha
```

```python
perplexities = []

for doc in df.intro_text:
    # preprocess document, break it into sentences, and convert to bigrams
    bigrams = process_document(doc)
    
    # Count bigram tokens in the sentence
    doc_counts = Counter()    
    doc_counts.update(bigrams)
    
    # Compute the perplexity of the document, given current Dirichlet prior alpha
    PP = compute_perplexity(bigrams, doc_counts, alpha)
    perplexities.append(PP)
    
    # Conjugate update to turn posterior into prior
    alpha = update_alpha(bigrams, alpha)
```

```python
perplexities = np.array(perplexities)
df_perplex_bayes = pd.Series(perplexities, index=df.index, name='Bayesian Perplexity')
```

```python
scaled_preplex_bayes = (df_perplex_bayes - df_perplex_bayes.mean()) / df_perplex_bayes.std()
```

3. We see a large change in perplexity in 2020. This is likely due to COVID, plus the change in ECB leadership. This is picked up in the similarity and in-sample perplexity metrics, but it is much more pronounced in the out-of-sample perplexity.

```python
from tqdm.notebook import tqdm

max_length = gpt2.config.n_positions
stride = 128
seq_len = encodings['input_ids'].size(1)
n_docs = encodings['input_ids'].size(0)

gpt2.to(device)
ppls = []

for i in tqdm(range(n_docs), total=n_docs, position=0, leave=True):
    prev_end_loc = 0
    nlls = []
    
    for begin_loc in range(0, seq_len, stride):
        end_loc = min(begin_loc + max_length, seq_len)
        trg_len = end_loc - prev_end_loc  # may be different from stride on last loop
        input_ids = encodings['input_ids'][i, begin_loc:end_loc].to(device)
        attention_mask = encodings['attention_mask'][i, begin_loc:end_loc].to(device)
        
        if torch.all(attention_mask == tokenizer.pad_token_id):
            break
            
        target_ids = input_ids.clone()
        target_ids[:-trg_len] = -100

        with torch.no_grad():
            outputs = gpt2(input_ids, labels=target_ids, attention_mask=attention_mask)

            # loss is calculated using CrossEntropyLoss which averages over input tokens.
            # Multiply it with trg_len to get the summation instead of average.
            # We will take average over all the tokens to get the true average
            # in the last step of this example.
            neg_log_likelihood = outputs.loss * trg_len

        nlls.append(neg_log_likelihood)

        prev_end_loc = end_loc
        if end_loc == seq_len:
            break

    ppl = torch.exp(torch.stack(nlls).sum() / end_loc)
    ppls.append(ppl.detach().cpu().numpy())
```

```python
gpt_perplexity = pd.Series(np.r_[ppls], index=df.index)
```

```python
scaled_gpt = (gpt_perplexity - gpt_perplexity.mean()) / gpt_perplexity.std()
```

In all cases, the Jaccard Similarity metric has a lowest variance in the period 1998-2008. This is quite interesting. I would expect a high variance in the out-of-sample Bayesian learning perplexity, because the model has a "burn in" period where it is adjusting the unconditional distribution over english words (the prior) to the conditional distribution over ECB documents. But even in the GPT2 model, which is incorporating information from all ECB documents across time (data leakage from the future), plus long-range dependencies within the documents, we see high variance in the perplexity from document to document. Evidently, the documents from this period were consistent with each other in some sense (stable Jaccard similarity), but 1) written in a way not at all familiar to someone who has not read ECB documents before, and 2) not consistently different from "normal" English documents.

There are two noteworthy differences between the GPT2 perplexity and all other metics. The first is the sharp drop in similarity in 2012, which is not present in any of the other similarlity measures. The second is a cyclical *decrease* in similarity between 2016-2019, while other models judge this period as increasing in similarlity. Finally is the sharp drop in 2021, which is captured by all models except the Bayesian learning model. The learning model puts the (COVID related? Accession of Christine Lagarde?) similarity drop in early 2020, which better matches my intuition about when the regime shift in communication should have occured, but the two "data leakage" models (MLE and GPT2) agree with the jaccard similarity metric, and put it in late 2020/early 2021.

## A bunch of regressions

Back to OLS because I'm a bit lazy. In principal this should be done Bayesian to allow real model comparisons.

For the life of me I have no idea why the authors do not include the non-interacted variables in the model. It seems utterly indefensible to suppose that the change in CAR, if it is effected by a change in sentiment at all, is thus effected *only* at the second order. So while I do as they do for model 1, I include the 1st order terms in my extension regressions. I also don't know why they take the log of the Jaccard similarity. They might have had the idea that this would make the data more normal, but since the similarity is bounded on the unit interval, *a region where the log function is extremely non-linear*, this has the effect of inducing a right-skew in the data. It would be much better to leave it as is, as the plot below shows. The log of the Jaccard similarity is manifestly non-normal, because it is bounded from above at 0. Yet more reasons to be Bayesian...

```python
fig, ax = plt.subplots(1, 2)
from scipy import stats

ax[0].plot(x_grid := np.linspace(0, 1, 1000), stats.gaussian_kde(data_df2.Similarity)(x_grid))
ax[1].plot(x_grid := np.linspace(-3, 1, 1000), stats.gaussian_kde(data_df2.Similarity.apply(np.log))(x_grid))
ax[0].set(title='Jaccard Similarity KDE')
ax[1].set(title='Log Jaccard Similarity KDE')
```

Model 1 is the same as the preferred specification in Amaya and Filbien (2015), with log Jaccard Similarity interacted with Pessimism. Results are identical to those presented above. Models 2-7 are extensions. Model 2 includes the level terms Pessimism and Similarity (in levels, not logs). We can see that both Pessimism and Similarity are significant (although at 10% and 5%), and the r2 has increased by 2% over Model 1. Clearly, there was no reason to drop these variables.

Models 3-7 consider different similarity specifications. In no other specification is Pessimism significant at the 10% level. It could be that including some notion of syntatic structure, even limited, also caputures something about the sentiment of the document. I.e., in bad times, there is a distributional shift in the way documents are written. This seems to make sense, but I'm reading way too far into a change in a 10% significant variable. If I had an 95% HDI over the posterior of Pessimism in Model 2, I would bet that zero would be included.

Likewise, the first-order similarity effect in all models is not significant, with the exception of GPT2, which is significant at the 5% level. 2nd order interaction is likewise not significant with any other similarity metric, including GPT2. Again, this suggests that something about using a language model absorbs the pass-through effect of sentiment on abnornal returns.

Also notable about model 4 (the GPT2 model) is that the output gap becomes significant at the 5% level. I don't have a theory for why this is the case, and of course it could be noise. If it is not noise, it must be that the richer language model is capturing some excluded mediator (in Judea Pearl's nomenclature) that is correlated with the output gap. This could be a more accurate measure of economic sentiment, or something else entirely.

Finally, since I observed that the GPT2 sentiment and the Bayesian sentiment are uncorrelated over the sample period, model 7 includes both. The GPT2-Similarity estimate becomes significant at the 1% level, and interactions between Pessimism and both Bayes Similarity and GPT2 similarity are significant. My first hypothesis about this was that data leakage (which Jaccard similarity does not suffer from, recall) might somehow be involved in mediating the interaction between similarity and sentiment. But this is rejected, because the interaction term in model 3 is not significant. 

Ideally, model 7 would also include the 3-way interaction Pessimism:Bayes_Similarity:GPT2_Similarity, but I don't have enough degrees of freedom. More reasons to be Bayesian.

The output_gap remains significant at the 5% level in model 7, so whatever effect GPT2_Similarity is absorbing, it does not seem to be correlated with Bayes_Similarity.
